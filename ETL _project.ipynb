```python
###Use Pandas and SQL Databases 
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
###EXTRACT: Store CSVs into DataFrames
#First Extract the data (as csv)
```

```python
grammy_data = "Resources/the_grammy_awards.csv"
spotify_data = "Resources/SpotifyFeatures.csv"
```

```python
#Create Dataframe from each CSV file
```

```python
grammy_data_df = pd.read_csv(grammy_data)
grammy_data_df.head(25)
```

```python
###TRANSFORM :Create new dataframes w select columns
#Renamed columns (must be lowercase and no spaces for PgAdmin)
#Dropped ALL NA and Dupicate song titles and artists to clean up DF
```

```python
grammy_columns = ["nominee", "artist", "year", "title", "category"]
grammy_data = grammy_data_df[grammy_columns].copy()

grammy_df = grammy_data.rename(columns={"nominee": "song_title",
                                        "artist": "artist_name",
                                        "year": "award_year",
                                        "title": "award_title",
                                        "category": "award_category"})

grammy_df = grammy_df.dropna(how='any')
grammy_df = grammy_df.drop_duplicates(subset=["song_title", "artist_name"], inplace=False)


grammy_df
```

```python
grammy_song_info_df = grammy_df[["song_title", "artist_name"]].copy()




grammy_song_info_df.head()
```

```python
award_info_df = grammy_df[["award_title", "award_year", "song_title", "artist_name", "award_category"]].copy()
award_info_df.head()
```

```python
spotify_data_df = pd.read_csv(spotify_data)
spotify_data_df.head(50)
```

```python
### Clean DataFrame
#Create new dataset w select columns
#Drop duplicates
```

```python
spotify_columns = ["artist_name", "track_name", "popularity", "danceability", "energy", "acousticness", "instrumentalness", "track_id", "key"]
spotify_data = spotify_data_df[spotify_columns].copy()

spotify_df = spotify_data.drop_duplicates(subset=["track_name", "artist_name"])

spotify_df = spotify_df.rename(columns={"track_name": "song_title"})



spotify_df
```

```python
spotify_song_info_df = spotify_df[["song_title", "artist_name"]].copy()




spotify_song_info_df.head()
```

```python
merged_df = grammy_song_info_df.append(spotify_song_info_df)
cleaned_merged_df = merged_df.drop_duplicates(subset=["song_title", "artist_name"])
cleaned_merged_df
```

```python
###Connect to local database (using pgAdmin - Postgres)
#Create Userstring and Password
#Check port (via properties in pgAdmin)
```

```python
rds_connection_string = "postgres:bootcamp@localhost:5432/music_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
###Check for tables - tables must have same names and be in the same order
#Create Engine
#create Database
#create SQL file
```

```python
engine.table_names()
```

```python
### LOAD Use Pandas to load DataFrames into database

```

```python
cleaned_merged_df.to_sql(name='song_info', con=engine, if_exists='append', index=False)
```

```python
award_info_df.to_sql(name='award_info', con=engine, if_exists='append', index=False)
```

```python
spotify_df.to_sql(name='spotify_data', con=engine, if_exists='append', index=False)
```
